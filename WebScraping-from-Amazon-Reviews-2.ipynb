```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_soup(u):
    r = requests.get(u,headers=HEADERS)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
review_list=[]
# soup
```

```python

# Function 2: look for web-tags in our soup, then append our data to reviewList
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'title': item.find('a', {'class': 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'}).text.strip(),
            'rating':float(item.find('i', {"data-hook":"review-star-rating"}).text.replace('out of 5 stars', '').strip()),
            'body': item.find('div', {'class': 'a-row a-spacing-small review-data'}).text.strip(),
            }
            review_list.append(review)
    except:
        pass
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) \
            Chrome/90.0.4430.212 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
for x in range(0,500):
    soup = get_soup(f'https://www.amazon.in/Samsung-Galaxy-M12-Storage-Replacement/product-reviews/B08XJG8MQM/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={x}')
    #print(soup)
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(review_list))
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break
```
